# Steganography

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### input carrier image

```python
img = cv2.imread('img1.jpeg')
img = np.array(img)
img[:,:,0],img[:,:,2] = np.array(img[:,:,2]), np.array(img[:,:,0])
plt.imshow(img)
```

### input sensitive image

```python
img2 = cv2.imread('img2.jpeg')
img2 = np.array(img2)
img2[:,:,0],img2[:,:,2] = img2[:,:,2],img2[:,:,0].copy()
plt.imshow(img2)
```

```python
print(img.shape)
print(img2.shape)
```

### function for steganography

```python
def steganography(img, img2):
    for k in range(3):
        for i in range(img2.shape[0]):
            col = 0
            for j in range(img2.shape[1]):
                c = 0
                while c != 8:
                    if img[i][col][k] % 2:
                        img[i][col][k] -= 1
                    img[i][col][k] += img2[i][j][k] % 2
                    img2[i][j][k] //= 2
                    col = col + 1
                    c = c + 1
    return img          
```

```python
steg = img.copy()
to_hide = img2.copy()
steg = steganography(steg,  to_hide)
```

### saving image for transfering

```python
plt.imshow(steg)

steg2 = steg.copy()
steg2[:, :, 0], steg2[:, :, 2] = steg2[:, :, 2], steg2[:, :, 0].copy()
# to avoid compression
cv2.imwrite('STEGANOGRAPHY.png', steg2, [0])
```

### function for extracting image

```python
def extraction(steg, shape):
    ext = np.zeros(shape, dtype = int)
    for k in range(3):
        for i in range(shape[0]):
            for j in range(shape[1] * 8):
                if j % 8 == 0:
                    pro = 1
                else:
                    pro *= 2
                ext[i][j // 8][k] += (steg[i][j][k] % 2) * pro
    return ext
```

```python
steg_img = cv2.imread('STEGANOGRAPHY.png', cv2.IMREAD_COLOR)
steg_img = np.array(steg_img)
steg_img[:, :, 0], steg_img[:, :, 2] = steg_img[:, :, 2], steg_img[:, :, 0].copy()
plt.imshow(steg_img)
```

```python
# Extracting from steganographic image
extract = extraction(steg_img, img2.shape)
```

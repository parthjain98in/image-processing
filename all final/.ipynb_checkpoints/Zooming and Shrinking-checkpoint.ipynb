```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("zoomnew.jpg", cv2.IMREAD_COLOR)
img[:,:,0], img[:,:,2] = img[:,:,2], img[:,:,0].copy()
```

```python
out = np.zeros((img.shape[0] * 2, img.shape[1] * 2, img.shape[2]), dtype=np.uint8)
```

## Zooming

```python
for i in range(img.shape[0]):
    out[2 * i, ::2] = np.copy(img[i, :])
```

```python
for k in range(out.shape[2]):
    for i in range(img.shape[0] - 1):
        for j in range(out.shape[1]):
            
            out[2 * i + 1][j][k] = out[2 * i][j][k] // 2 + out[2 * (i + 1)][j][k] // 2
```

```python
for k in range(out.shape[2]):
    for i in range(out.shape[0]):
        for j in range(img.shape[1] - 1):
            out[i][2 * j + 1][k] = out[i][2 * j][k] // 2 + out[i][2 * (j + 1)][k] // 2
```

```python
out[-1, :, :] = np.array(out[-2, :, :])
out[:, -1, :] = np.array(out[:, -2, :])
```

```python
fig = plt.figure(figsize = (20, 20))

ax = fig.add_subplot(121)
ax.imshow(img)
ax.set_title('Original')

bx = fig.add_subplot(122)
bx.imshow(out)
bx.set_title('Zoomed')

plt.show()
```

## Shrinking

```python
out_s = np.array(img[::2, ::2, :])
```

```python
fig = plt.figure(figsize = (20, 20))

ax = fig.add_subplot(121)
ax.imshow(img)
ax.set_title('Original')

bx = fig.add_subplot(122)
bx.imshow(out_s)
bx.set_title('Shrinked')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("mona lisa.jpg", cv2.IMREAD_COLOR)
image[:, :, 0], image[:, :, 2] = np.array(image[:, :, 2]), np.array(image[:, :, 0])
```

```python
plt.imshow(image)
```

```python
shape = image.shape
print(shape)
```

Red

```python
red_image = np.array(image)
red_image[:, :, 1:] = 0
plt.imshow(red_image)
```

Green

```python
green_image = np.array(image)
green_image[:,:,0] = 0
green_image[:,:,2] = 0
plt.imshow(green_image)
```

Blue

```python
blue_image = np.array(image)
blue_image[:,:,0:2] = 0
plt.imshow(blue_image)
```

Gray

```python
gray_image = np.zeros(shape[:2])
for i in range(shape[0]):
    for j in range(shape[1]):
        r = image[i][j][0]
        g = image[i][j][1]
        b = image[i][j][2]
        gray_image[i][j] = 0.3 * r + 0.59 * g + 0.11 * b
```

```python
plt.imshow(gray_image, cmap="gray")
```
